```python
#converting to datetime datatype

d['date']=pd.to_datetime(d['date'])
d.info()
```

```python
d['date_year']=d['date'].dt.year
d['date_month_no']=d['date'].dt.month
d['date_month_name']=d['date'].dt.month_name()
d['date_day']=d['date'].dt.day
d['date_dow']=d['date'].dt.dayofweek
d['date_dow_name']=d['date'].dt.day_name()

d.drop(columns=['product_id','city_id','orders']).head(10)
```

```python
d['date_is_weekend']=np.where(d['date_dow_name'].isin(['Sunday','Saturday']),1,0)
```

```python
d.head()
```

```python
d['date_week'] = d['date'].dt.isocalendar().week
d['quarter']=d['date'].dt.quarter
```

```python
d.head()
```

```python
d['semester']=np.where(d['quarter'].isin([1,2]),1,2)
d.drop(columns=['product_id','city_id','orders']).head()
```
