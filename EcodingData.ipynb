Ordinal Encoding

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\KIIT0001\Downloads\customer.csv")
```

```python
df.sample(5)
```

```python
df=df.iloc[:,2:]
```

```python
df.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(
    categories=[
        ['Poor', 'Average', 'Good'],  # Feature 1 order
        ['School', 'UG', 'PG']            # Feature 2 order
    ]
)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.iloc[:,0:2],df.iloc[:,-1],test_size=0.2)
```

```python
X_train
```

```python
oe.fit(X_train)
```

```python
X_train=oe.transform(X_train)
X_test=oe.transform(X_test)
```

```python
X_train
```

```python
oe.categories_
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.model_selection import train_test_split

X = df.iloc[:, 0:4]   # brand, km_driven, fuel, owner
y = df.iloc[:, -1]    # selling_price

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=2
)
```

```python
import numpy as np

ohe = OneHotEncoder(
    drop='first',
    sparse_output=False,   # ✅ new parameter
    dtype=np.int32
)
```

```python
# fit only on training data
X_train_new = ohe.fit_transform(X_train[['fuel','owner']])

# transform test data
X_test_new = ohe.transform(X_test[['fuel','owner']])
```
