```python
import numpy as np
import pandas as pd
```

```python
df=df=pd.read_csv(r"C:\Users\KIIT0001\Downloads\titanic.csv")
```

```python
df.sample(10)
```

```python
df['number'].unique()
```

```python
fig=df['number'].value_counts().plot.bar()
fig.set_title('Passenger travelling with')
```

```python
df['number_numerical']=pd.to_numeric(df["number"],errors='coerce',downcast='integer')
```

```python
df['numeric_categorical']=np.where(df['number_numerical'].isnull(),df['number'],np.nan)
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

```python
df['cabin_num']=df['Cabin'].str.extract('(/d+)')
df['cabin_cat']=df['Cabin'].str[0]

df.head()
```

```python
fig=df["cabin_cat"].value_counts().plot.bar()
```

```python
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(
    df['ticket_num'],
    errors='coerce',
    downcast='integer'
)

df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(
    df['ticket_cat'].str.isdigit(),
    np.nan,
    df['ticket_cat']
)

df.head(20)
```
